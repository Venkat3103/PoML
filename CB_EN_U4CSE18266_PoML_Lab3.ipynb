### Import necessary packages

```python
import numpy as np
import pandas as pd
```

### Import dataset

```python
dataset = pd.read_csv('practice_data.csv')
```

### How many columns in the dataset?

```python
print("No of Columns in the Dataset:",dataset.shape[1])
```

### Identify the categorical variables

```python
CVCols = dataset.select_dtypes(include=['object'])
print("Categorical Variables:", CVCols.columns.tolist())
```

### How many columns have missing values?

```python
dataset.isnull().sum()
```

### Replace missing values of categorical variables with mode

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
CV = CVCols.values
imputer.fit(CV[:, :])
CV[:, :] = imputer.transform(CV[:, :])
for column in CVCols:
    dataset[column] = CVCols[column]
dataset
```

### Replace missing values of age column with median

```python
Age = dataset['Age'].values.reshape(-1, 1)
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(Age[:,:])
Age = imputer.transform(Age[:,:])
dataset['Age'] = Age.flatten()
dataset
```

### Replace missing values of salary with mean salary

```python
Salary = dataset['Salary'].values.reshape(-1, 1)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(Salary[:,:])
Salary = imputer.transform(Salary[:,:])
dataset['Salary'] = Salary.flatten()
dataset
```

### In case of any NaN value in response column, delete the entire instance

```python
dataset = dataset[pd.notnull(dataset['Response'])]
dataset
```

### Find the range of age

```python
print("Age ranges from",dataset['Age'].min(),"to",dataset['Age'].max())
```

### Apply min-max scaling on salary to transform it into (0,1)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
dataset['Salary'] = min_max_scaler.fit_transform(dataset['Salary'].values.reshape(-1,1)).flatten()
dataset
```
