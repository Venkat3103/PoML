```
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Venkat3103/PoML/main/Salaries.csv",na_values="Not Provided")
```

1. How many records are available in the dataset?

```
print("Number of records: ",df.shape[0])
```

2. Find the total number of attributes and their name.

```
print("Total number of attributes:",df.shape[1])
print("Names of attributes: ",df.columns.tolist())
```

3. Find the average total salary

```
print("Average Total Salary",df['TotalPay'].mean())
```

4. List the employee names whos total salary is above the average

```
print("Employees with salary greater than average salary: ",df[df.TotalPay>df.TotalPay.mean()]["EmployeeName"].tolist())
```

5. Find the names of employees whoes base pay belongs to 75-percentile of the attribute

```
print("Employees whos base pay belongs to 75-percentile",df[df.TotalPay>df.describe().TotalPay['75%']].EmployeeName.tolist())
```

6. Find the average basepay in year wise, which year have highest base pay

```
print("Year with highest average basepay: ",df.groupby('Year')['BasePay'].mean().sort_values().index[-1])
```

7. Plot the average totalpay vs year (using matplotlib)

```
import matplotlib.pyplot as plt
plt.figure(figsize=(4,5))
b=df.groupby('Year')['TotalPay'].mean()
plt.title("Yearwise Average Salary")
plt.bar(b.index,b.values);
plt.ylim(70000,80000)
```
